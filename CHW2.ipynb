A structure's various components can be represented by a single **state vector**, which changes over time due to certain processes. This changing state vector, when modeled using linear transformations at discrete intervals, forms the basis of a **discrete dynamical system**.

Starting with an initial state $X_0$, transitions to subsequent states are governed by a matrix $P$, embodying linear transformations. For instance, $X_1 = PX_0$ marks the first transition, leading to a sequence where $X_2 = P^2X_0$, $X_3 = P^3X_0$, and so forth, generalizing to $X_t = PX_{t-1}$ for each time step $t$.

This approach is particularly effective in modeling web navigation, where each webpage's visit probability at any given time is captured by the state vector $X_t$. The transition matrix $P$ specifies the likelihood of moving from one page to another, with each element $p_{ij}$ representing the probability of transitioning from page $j$ to page $i$. To ensure the model's consistency, the sum of probabilities in each column of $P$ must equal one, reflecting the comprehensive distribution of navigating from any given page to the next.

```python
## IMPORTS 
import numpy as np
import math
import numpy as np

def discrete_dynamical_system(P, X0, num_steps):
    """
    Simulates a discrete dynamical system and models web navigation using a Markov chain.
    
    Parameters:
    P (numpy.ndarray): Transition matrix
    X0 (numpy.ndarray): Initial state vector
    num_steps (int): Number of time steps to simulate
    
    Returns:
    numpy.ndarray: State vectors at each time step
    """
    # Check if the transition matrix is valid
    if not np.allclose(np.sum(P, axis=0), 1):
        raise ValueError("The transition matrix is not valid. The sum of probabilities in each column must equal 1.")
    
    # Initialize the state vectors
    states = np.zeros((num_steps+1, len(X0)))
    states[0] = X0
    
    # Simulate the discrete dynamical system
    for t in range(1, num_steps+1):
        states[t] = states[t-1] @ P
    
    return states

# Example usage
# Define the transition matrix
P = np.array([[0.5, 0.2, 0.3],
              [0.1, 0.6, 0.3],
              [0.2, 0.3, 0.5]]).T

# Define the initial state vector
X0 = np.array([0.4, 0.3, 0.3])

# Simulate the system for 5 time steps
num_steps = 5
states = discrete_dynamical_system(P, X0, num_steps)

# Print the state vectors at each time step
for t in range(num_steps+1):
    print(f"X_{t} = {states[t]}")
```

In a more realistic model of the web, some pages may not have outbound links (dangling nodes), and users may sometimes choose to navigate to a random page rather than following an outbound link. The Google PageRank algorithm introduces a damping factor to model this behavior, making the Markov process more reflective of real-world web navigation.

Dangling nodes are pages without outbound links. These nodes would lead to columns of zeros in the transition matrix, which is problematic for calculating the steady state. To address this, distribute the probability mass of dangling nodes equally among all pages.

First identify dangling nodes then djust the transition matrix $P$ by redistributing the dangling nodes' probability mass equally across all pages, ensuring every column sums to 1.

```python
import numpy as np

# Example transition matrix for a small web model, including a dangling node
# Assume 4 pages, with the 4th page being a dangling node
P = np.array([
    [0, 0, 0, 0, 0],     # Probabilities from each page to Page 1
    [1, 0, 0.5, 0, 0],   # Probabilities from each page to Page 2
    [0, 0, 0, 0.5, 0],   # Probabilities from each page to Page 3
    [0, 1, 0.5, 0, 0],   # Probabilities from each page to Page 4
    [0, 0, 0, 0.5, 0]    # Page 5 is a dangling node
])

### START CODE HERE ###

# Identify dangling nodes (columns with all zeros)
dangling_nodes = np.where(~P.any(axis=0))[0]

# Adjust the transition matrix by redistributing the probability mass of dangling nodes equally
num_pages = P.shape[0]
equal_prob = 1 / num_pages

for node in dangling_nodes:
    P[:, node] = equal_prob

# Ensure every column sums to 1
P = P / np.sum(P, axis=0)

### END CODE HERE ###

print("Adjusted Transition Matrix P:")
print(P)
```

The damping factor models the behavior of web surfers who occasionally jump to a random page instead of following outbound links. This factor, usually set to 0.85, balances between following links ($P$) and jumping at random ($\frac{1}{n}$, where \(n\) is the number of pages).

Apply the damping factor $d$ to the transition matrix $P$ to get $P' = dP + \frac{1-d}{n}E$, where $E$ is an $n \times n$ matrix with all elements equal to $\frac{1}{n}$.

```python

# Define the damping factor d
d = 0.85
n = P.shape[0]

# Create the E matrix
E = np.full((n, n), 1/n)

# Apply the damping factor to the transition matrix
P_prime = d * P + (1 - d) * E

print("Adjusted Transition Matrix P':")
print(P_prime)
```

$$
\hat{y}[t] = \sum_{i=1}^{p} a_i y[t-i]
$$, 
where $\hat{y}[t]$ is the predicted value at time $t$, $y[t-i]$ are the past values, and $a_i$ are the coefficients that need to be determined. Construct an embedding matrix from lagged versions of the data series. This matrix, often denoted as $M$, captures the structure of dependencies in past values. Use the pseudo-inverse of $M$ to solve for the coefficients $a_i$ that minimize the prediction error in a least squares sense.

To construct an embedding matrix from lagged versions of the data series and solve for the coefficients $$a_i$$ that minimize the prediction error in a least squares sense, you can follow these steps:

1. Create the embedding matrix $M$ by stacking lagged versions of the data series. Each row of $M$ corresponds to a time point, and each column corresponds to a lag. For example, if you have a data series $\{y[t]\}$ and you want to use the last p values to predict the current value, the embedding matrix would be:

$$
M = \begin{bmatrix}
y[1] & y & 0 \\
y[2] & y[1] & y \\
\vdots & \vdots & \vdots \\
y[T] & y[T-1] & y[T-2]
\end{bmatrix}
$$

where $T$ is the length of the data series and the missing values are padded with zeros.

2. Let $\mathbf{y}$ be the vector of target values to predict, where $y[t]$ is the target for time $t$. The goal is to find coefficients $\mathbf{a} = [a_1, a_2, a_3]^T$ that minimize the mean squared error between the predicted values $\hat{\mathbf{y}}$ and the actual values $\mathbf{y}$:

$$
\mathbf{a} = \arg\min_{\mathbf{a}} \|\mathbf{y} - M\mathbf{a}\|_2^2
$$

3. The solution to this least squares problem is given by:

$$
\mathbf{a} = (M^T M)^{-1} M^T \mathbf{y}
$$

where $M^T$ is the transpose of $M$, and $$(M^T M)^{-1} M^T$$ is the pseudo-inverse of $M$.

4. Once you have the coefficients $\mathbf{a}$, you can use them to predict future values of the time series using the autoregressive equation:

$$
\hat{y}[t] = \sum_{i=1}^{p} a_i y[t-i]
$$


```python
import numpy as np

# Generate sample data
data = data_array[0].T
p = 100  # Number of lags

# Construct the embedding matrix M
M = np.column_stack([np.roll(data, i) for i in range(p, 0, -1)])

# Target values to predict
y = data

# Calculate the coefficients using the pseudo-inverse
coefficients = np.linalg.pinv(M) @ y

# Make predictions
predictions = M @ coefficients

# Print the coefficients and predictions
print("Coefficients (a_i):", coefficients)
print("Predictions:", predictions)
```

Assess the fitted model by comparing the predicted values against actual values using metrics such as the Mean Squared Error (MSE) or the coefficient of determination ($R^2$).

```python

from sklearn.metrics import mean_squared_error, r2_score



# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y, predictions)
print("Mean Squared Error (MSE):", mse)

# Calculate the coefficient of determination (R^2)
r2 = r2_score(y, predictions)
print("Coefficient of Determination (R^2):", r2)
```

To formulate the inverse problem as a least squares problem based on the given scenario, where we flatten the grayscale image into a vector, apply a filter by multiplying $L$ with the flattened image vector ($x_{filtered} = Lx$), and then reshape $x_{filtered}$ back to the original image dimensions, we can proceed as follows:

### Theoretical Formulation of the Inverse Problem as a Least Squares Problem:

   - $\mathbf{x}$: Flattened vector representing the original image.
   - $\mathbf{L}$: Filter matrix.
   - $\mathbf{x_{filtered}}$: Result of applying the filter, obtained by $\mathbf{x_{filtered}} = \mathbf{Lx}$.
   - $\mathbf{d}$: Observed data, which is $\mathbf{x_{filtered}}$ in this case.

   The least squares problem seeks to minimize the squared difference between the observed data and the predicted data:
   
   $\min_{\mathbf{x}} \|\mathbf{d} - \mathbf{Gx}\|^2$
   
   In this case, $\mathbf{G}$ is the filter matrix $\mathbf{L}$.

   - $\min_{\mathbf{x}} \|\mathbf{x_{filtered}} - \mathbf{Lx}\|^2$


By formulating the inverse problem as a least squares problem, we can apply optimization techniques to find the best estimate of the original image vector that fits the observed filtered image data.
